```python
import pandas as pd
import numpy as np
```

```python
Bee2015_Jan_March_to_load = "Jan-Mar2015.csv"

# Read data File and store into Pandas Data Frames
Jan_March2015= pd.read_csv(Bee2015_Jan_March_to_load)
Jan_March2015.head()
```

```python
Jan_March2015_df = Jan_March2015.drop(Jan_March2015.columns[[0, 1]], axis=1) 

Jan_March2015_df.head()
```

```python
list(Jan_March2015_df)
```

```python
Jan_March2015_Final= Jan_March2015_df.rename(columns={'State':'State',
                                                      'Number of colonies': 'Winter_2015\nNo_of_colonies', 
                                                      'Maximum colonies ': 'Winter_2015\nMax',
                                                     'Lost colonies': 'Winter_2015\nLost', 
                                                      'Percent lost ': 'Winter_2015\nPer_lost',
                                                         'Added colonies': 'Winter_2015\nAdded',
                                                      'Renovated colonies ': 'Winter_2015\nRenovated', 
                                                     'Percent renovated ': 'Winter_2015\nPer_renovated'})

Jan_March2015_Final.head()
```

```python
Jan_March2015_Final.replace('-',np.NaN,inplace = True)
Jan_March2015_Final.replace('(Z)',np.NaN,inplace = True)

Jan_March2015_Final.fillna(0,inplace = True)

Jan_March2015_Final.head()
```
