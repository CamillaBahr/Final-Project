```python
import pandas as pd
import numpy as np
```

```python
Bee2015_Stressors_Jan_March_to_load = "Jan-Mar2015_Stressors.csv"

# Read data File and store into Pandas Data Frames
Jan_March2015_Stressors = pd.read_csv(Bee2015_Stressors_Jan_March_to_load)

Jan_March2015_Stressors.head()
```

```python
Jan_March2015_Stressors_df = Jan_March2015_Stressors.drop(Jan_March2015_Stressors.columns[[0, 1]], axis=1)

Jan_March2015_Stressors_df.head()
```

```python
list(Jan_March2015_Stressors_df)
```

```python
Jan_Mar2015_Stressors_Final= Jan_March2015_Stressors_df.rename(columns={'State':'State',
                                                                            'Varroa mites': 'Winter_2015\nMites', 
                                                                           'Other pests and parasites ': 'Winter_2015\nParasites',
                                                                             'Diseases ': 'Winter_2015\nDiseases', 
                                                                            'Pesticides': 'Winter_2015\nPesticides',
                                                                             'Other': 'Winter_2015\nOther', 
                                                                            'Unknown': 'Winter_2015\nUnknown'})

Jan_Mar2015_Stressors_Final.head(10)
```

```python
Jan_Mar2015_Stressors_Final.replace('-',np.NaN,inplace = True)
Jan_Mar2015_Stressors_Final.replace('(Z)',np.NaN,inplace = True)

Jan_Mar2015_Stressors_Final.fillna(0,inplace = True)

Jan_Mar2015_Stressors_Final.head()
```
