```python
import pandas as pd
import numpy as np
```

```python
SN_Bee2016_Jan_March_to_load = "Jan-March2016.csv"

# Read data File and store into Pandas Data Frames
Jan_March2016= pd.read_csv(SN_Bee2016_Jan_March_to_load)

# Combine the data into a single dataset
#school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
Jan_March2016.head()
```

```python
Jan_March2016_Final= Jan_March2016_df.rename(columns={'No_of_colonies': 'Winter_2016\nNo_of_colonies', 
                                                           'Max': 'Winter_2016\nMax',
                                 'Lost': 'Winter_2016\nLost', 'Per_lost': 'Winter_2016\nPer_lost',
                                 'Added': 'Winter_2016\nAdded', 'Renovated': 'Winter_2016\nRenovated', 
                                 'Per_renovated': 'Winter_2016\nPer_renovated'})

Jan_March2016_Final.head()
```
